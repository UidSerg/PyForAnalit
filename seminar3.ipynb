```python
import pandas as pd
df = pd.read_csv('laptop_price.csv', encoding='windows-1258', sep=",")
df.head()
```

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda x: x.split(" ")[0])
df.head()
```

```python
df['Memory_Amount'] = df['Memory'].apply(lambda x: x.split(" ")[0][:-2])
df.head()
```

```python
def memory(memory_str):
    gb=memory_str.split(" ")[0][:-2]
    if gb.endswith('GB'):
       gb=float(gb[:-2]) 
    elif gb.endswith('TB'):
        gb=float(gb=gb[:-2])*1024
    return gb
df['Memory_Amount'] = df['Memory'].apply(memory)
df.head()
```

```python
df['Memory_type'] = df['Memory'].apply(lambda x: x.split(" ")[1])
df.head()
```

1.4 Удалите признаки Memory и ScreenResolution


```python
df.drop(columns=['Memory', 'ScreenResolution'], inplace=True)
df.head()
```
