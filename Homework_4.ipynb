1.1. Скачать следующие данные: kc-house-data и laptop_price

```python
import pandas as pd
df = pd.read_csv('kc_house_data.csv', encoding='windows-1258', sep=",")
df.head()
```

```python
import matplotlib.pyplot as plt
```

1.2. Изучите стоимости недвижимости

```python
plt.figure(figsize=(5,8))
plt.hist(df['price'])
plt.title("График по цене")
plt.xlabel("цена")
plt.ylabel("количество")
plt.xticks(rotation=-90)
```

> Больще всего домов стоимостью ~ до 1 миллиона

1.3. Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(5,8))
plt.hist(df['sqft_living'])
plt.title("График квадратуре жилья")
plt.xlabel("квадраты")
plt.ylabel("количество")
plt.xticks(rotation=-90)
```

> Больще всего домов с площадью около 2000 квадратных метров

1.4. Изучите распределение года постройки

```python
plt.figure(figsize=(8,4))
plt.hist(df['yr_built'])
plt.title("График по годам постройки")
plt.xlabel("годы")
plt.ylabel("количество")
plt.xticks(rotation=-90)
```

> с каждым годом жилья строилось все больше

```python
data = df['waterfront'].value_counts()
plt.pie(data.values, autopct='%.1f%%', labels=data.index)
plt.title("График домов с видом на набережную ")
plt.legend(["нет", "есть"])
```

```python
data1 = df['floors'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(data1.values, autopct='%.1f%%', labels=data1.index)
plt.title("Этажность домов")
plt.legend(data1.index)
```

> Почти половина из всех домов одноэтажные. В оснвоном дома одно и 2х этажные

2.3. Изучите распределение состояния домов

```python
data3 = df['condition'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(data3, autopct='%1.1f%%', labels=data3.index);
plt.legend(data3.index)
```

```python
import numpy as np
```

Матрица корреляций

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(10, 8))
sns.set(font_scale=1.4)
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='RdYlGn')
plt.title('Correlation matrix');
```
